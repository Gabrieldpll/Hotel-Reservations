# Transformação de variável temporal

**Modelagem da variável temporal**

Transformamos a variável temporal em trimestres e um indicador se é feriado.

```python
def transforma_trimeste(x):
  if x in [1,2,3]:
    return 'primeiro_trimestre'
  elif x in [4,5,6]:
    return 'terceiro_trimestre'
  elif x in [7,8,9]:
    return 'terceiro_trimestre'
  elif x in [10,11,12]:
    return 'quarto_trimestre'

dados['Trimestre'] = dados['arrival_month'].map(transforma_trimeste)
```
